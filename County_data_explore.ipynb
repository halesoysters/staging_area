```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = Path('Christina_county_pop.csv')
cc_df = pd.read_csv(data, skiprows = 1)
cc_df
```

```python
cc_df.dtypes
```

```python
#create a new column the classifies countries os Rural, micro urban or Urban based on total population
#create a list of conditions
conditions = [(cc_df['ORG_DENSITY']<=500),(cc_df['ORG_DENSITY'] >500)]

#create the list of values to assign for each condition
values = ['Rural', 'Urban']

#create new column and use np.select to assign values
cc_df['RuralUrban'] = np.select(conditions, values)

cc_df
```

```python
#display number of rural vs urban counties
cc_df['RuralUrban'].value_counts()
```

```python
#drop empty columns
cc_short = cc_df.drop(columns=['Geo_FIPS', 'Geo_NATION', 'Geo_STATE', 'Geo_COUNTY',
                               'Geo_CT', 'Geo_BG', 'Geo_CITY', 'Geo_MSAMSA', 'Geo_ZC'])

cc_short
```

# Importing county data to join

```python
data = Path('county_pop_df.csv')
county_pop_df = pd.read_csv(data)
county_pop_df
```

```python
county_pop_df['Geo_QNAME'] = county_pop_df['CountyName'].astype(str) + ', ' + county_pop_df ['StateDesc']

county_pop_df
```

```python
county_pop_df.drop(columns=['RuralUrban'], inplace = True)

county_pop_df
```

```python
#merge DataFrames
merged_df = pd.merge(county_pop_df, cc_short, how='outer', on=['Geo_QNAME'])

merged_df
```

```python
#find null values
for column in merged_df.columns:
    print(f"Column {column} has {merged_df[column].isnull().sum()} null values")
```

```python
#after dropping na, same number of rows in DF
merged_df.dropna().count()
```

```python
merged_df.to_csv('County_Merged.csv', index=False)
```
